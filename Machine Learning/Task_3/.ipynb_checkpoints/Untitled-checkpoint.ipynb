```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
import seaborn as sns
```

```python
matches_df = pd.read_csv('matches.csv')
```

```python
matches_df.shape
```

```python
matches_df.drop("Unnamed: 0",axis=1, inplace=True)
```

```python
list1 = [1, 13274, 15722, 17642, 19694, 24558]
for i in list1:
   matches_df.drop(matches_df[ matches_df['country_id'] == i ].index, inplace = True)
```

```python
matches_df.shape
```

```python
matches_df
```

```python
country_df = pd.read_csv('Country.csv')
country_df
```

```python
country_df.rename(columns={'id': 'country_id','name': 'country_name'}, inplace=True)
country_df
```

```python
league_df = pd.read_csv('league.csv')
league_df
```

```python
league_df.rename(columns={'name': 'league_name','name.1': 'country_name','id.1': 'league_id'}, inplace=True)
league_df.drop('id', axis=1, inplace = True)
```

```python
league_df
```

```python
league_country_df = pd.merge(country_df, league_df, on='country_id', how='outer')
league_country_df
```

```python
matches_df_t2 = matches_df[['country_id','league_id','season','match_api_id','home_team_api_id','away_team_api_id','home_team_goal','away_team_goal']].copy()
```

```python
matches_df_t2
```

```python
df = pd.merge(matches_df_t2, league_country_df, on='country_id', how='left').drop_duplicates()
```

```python
df
```

```python
df_team = pd.read_csv('Team.csv')
df_team.columns
df_team.shape
```

```python
team_attri = pd.read_csv('Team_Attributes.csv')
team_attri.columns
team_attri.shape
```

```python
df['Home_team_name'] = df['home_team_api_id'].map(df_team.set_index('team_api_id')['team_long_name'].drop_duplicates())
df
```

```python
df['Away_team_name'] = df['away_team_api_id'].map(df_team.set_index('team_api_id')['team_long_name'].drop_duplicates())
df
```

```python
df['goals'] = df['home_team_goal']+df['away_team_goal']
```

```python
df_q2_goal_sum = df.groupby(['season', 'country_name']) ['goals'].sum().reset_index()
df_q2_goal_sum.head()
```

```python
df_q2_avg_goals = df.groupby(['country_name', 'season'])['goals'].mean().reset_index()
df_q2_avg_goals
```

```python
plt.plot(df_q2_avg_goals[df_q2_avg_goals['country_name'] == 'England']['season'], df_q2_avg_goals[df_q2_avg_goals['country_name'] == 'England']['goals'], label = 'England')
plt.plot(df_q2_avg_goals[df_q2_avg_goals['country_name'] == 'Spain']['season'], df_q2_avg_goals[df_q2_avg_goals['country_name'] == 'Spain']['goals'], label = 'Spain')
plt.plot(df_q2_avg_goals[df_q2_avg_goals['country_name'] == 'Germany']['season'], df_q2_avg_goals[df_q2_goal_sum['country_name'] == 'Germany']['goals'], label =  'Germany')
plt.plot(df_q2_avg_goals[df_q2_avg_goals['country_name'] == 'Italy']['season'], df_q2_avg_goals[df_q2_avg_goals['country_name'] == 'Italy']['goals'], label = 'Italy')
plt.plot(df_q2_avg_goals[df_q2_avg_goals['country_name'] == 'France']['season'], df_q2_avg_goals[df_q2_avg_goals['country_name'] == 'France']['goals'], label = 'France')

plt.xlabel("Season")
plt.ylabel("Number of goals")

plt.legend(loc='upper right')
plt.show()
```
