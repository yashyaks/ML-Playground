```python
matches_df
```

```python
leagues_df = pd.read_csv('leagues.csv')
```

```python
leagues_df.drop("Unnamed: 0",axis=1, inplace=True)
```

```python
leagues_df.rename(columns={'name': 'league_name','name.1': 'country_name','id.1': 'league_id'}, inplace=True)
```

```python
leagues_df
```

```python
df = pd.merge(matches_df, leagues_df, on='id', how='left')
```

```python
df['Dates'] = pd.to_datetime(df['date']).dt.date
df['Time'] = pd.to_datetime(df['date']).dt.time
```

```python
df
```

```python
df_q3=df.groupby(['country_name'])  
```

```python
sns.lineplot(x = 'Dates', y = 'home_team_goal', data = df_q3)
```
