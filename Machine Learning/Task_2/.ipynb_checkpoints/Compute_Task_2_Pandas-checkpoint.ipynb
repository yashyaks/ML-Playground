## Filtering

```python
df_filt = df[df['duration_ms'] < 400000]
df_filt
```

```python
df_sort = df_filt.sort_values('energy')##[['song_title', 'artist', 'duration_ms']]
df_sort
```

```python
result = df_sort.sort_values('energy', ascending=False)
result
```

## Exporting a df to CSV

### after the above filtering and sorting we have made a df 'result' with the duration more than 400000ms sorted in descending order of dancebility

```python
result.to_csv(r'C:\Users\Yash Thakar\PROGRAMMING\compute_tasks\Machine Learning\Task_2\filtered_data.csv')
```

## Activities

### Find songs with value for energy column greater than 0.4 and display values for song_title column.

```python
df_filt2 = df[df['energy']>0.4]
display(df_filt2.loc[:,'song_title'])
```

### For songs with duration duration_ms value less than 300000 ms (300 s), find Top 10 “fastest” songs( songs with highest values for tempo column ) and display values for song_title, artist columns.



```python
df_filt3 = df[df['duration_ms']<300000]
df_sort2 = df_filt3.nlargest(10,'tempo')
display(df_sort2.loc[:,['song_title','artist']])
```

### Find songs with value for artist column = Drake and value for energy column greater than 0.5 and display values for song_title column.

```python
df_filt4 = df[(df['artist'] == 'Drake') & (df['energy']>0.5)]

display(df_filt4.loc[:,'song_title'])
```
